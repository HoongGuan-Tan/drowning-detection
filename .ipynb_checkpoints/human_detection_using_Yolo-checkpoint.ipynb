## Human Detection using YOLO

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load YOLO model
model = YOLO("yolo11n.pt")

# Read and preprocess image
img = cv2.imread('tester1.jpg')
img = cv2.resize(img, (640,640))

# Convert image to RGB for display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Run inference
results = model(img, conf=0.3)

# Get detected boxes
for result in results:
    boxes = result.boxes.xyxy  # Get bounding boxes (x1, y1, x2, y2)
    confs = result.boxes.conf  # Confidence scores
    cls_ids = result.boxes.cls  # Class IDs

    for box, conf, cls_id in zip(boxes, confs, cls_ids):
        if int(cls_id) == 0:  # Only 'person'
            x1, y1, x2, y2 = map(int, box)  # Convert box to int
            cv2.rectangle(img_rgb, (x1, y1), (x2, y2), (255, 0, 0), 2)  # Draw blue box
            cv2.putText(img_rgb, f"person {conf:.2f}", (x1, y1 - 5),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)  # Label

# Display only 'person' detections
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

## Image Cropping

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('tester1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Convert YOLO format bbox to absolute bbox
with open('tester1.txt', "r") as f:
    lines = f.readlines()
    
for line in lines:
    class_id, x_c, y_c, w, h = line.strip().split()
    x_c = float(x_c)
    y_c = float(y_c)
    w = float(w)
    h = float(h)

x1 = (x_c - w / 2)
y1 = (y_c - h / 2)
x2 = (x_c + w / 2)
y2 = (y_c + h / 2)

height, width = img.shape
xmin = int(x1 * width)
ymin = int(y1 * height)
xmax = int(x2 * width)
ymax = int(y2 * height)

# Crop Image
roi = img[ymin:ymax, xmin:xmax]
roi = cv2.resize(roi, (224,224))

# Save the Cropped Image
cv2.imwrite('cropped_img.jpg', roi)

# Show
cv2.imshow('img', roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
