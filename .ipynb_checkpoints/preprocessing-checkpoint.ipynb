## Import Libraries

```python
import os
import csv
import cv2
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import shutil
import random
from tqdm import tqdm
```

## Preprocessing

The original dataset consists of large images and bounding box labels in YOLO format \
In this part, we make use of the bounding box labels to `identify the roi and crop the images` \
Then we `resize the images to 32x32` \
The processed images are saved in new directory

 - processed_dataset \
    - train \
        - swimming \
        - tread_water \
        - drowning \
    - val \
        - swimming \
        - tread_water \
        - drowning \

```python
def process_dataset(input_base_path, output_base_path):
    # Class mapping for reference
    class_map = {
        '0': 'swimming',
        '1': 'tread_water',
        '2': 'drowning'
    }
    
    # Create output directories for each class
    for dataset_type in ['train', 'val']:
        for class_name in class_map.values():
            os.makedirs(os.path.join(output_base_path, dataset_type, class_name), exist_ok=True)
    
    # Initialize counter for sequential naming (separate counter for each class)
    class_counters = {
        'swimming': 0,
        'tread_water': 0,
        'drowning': 0
    }
    
    # Process training data first, then validation data
    for dataset_type in ['train', 'val']:
        print(f"Processing {dataset_type} dataset...")
        
        # Get all image paths
        image_folder = os.path.join(input_base_path, 'images', dataset_type)
        label_folder = os.path.join(input_base_path, 'labels', dataset_type)
        
        image_paths = sorted(glob.glob(os.path.join(image_folder, '*.jpg')))
        
        # Process each image
        for image_path in tqdm(image_paths):
            # Get corresponding label file
            image_name = os.path.basename(image_path)
            label_name = os.path.splitext(image_name)[0] + '.txt'
            label_path = os.path.join(label_folder, label_name)
            
            # Skip if label file doesn't exist
            if not os.path.exists(label_path):
                print(f"Warning: No label found for {image_name}")
                continue
            
            # Read image
            img = cv2.imread(image_path)
            height, width = img.shape[:2]
            
            # Read labels
            with open(label_path, 'r') as f:
                lines = f.readlines()
            
            # Process each bounding box
            for line in lines:
                parts = line.strip().split()
                
                # Skip if the format is incorrect
                if len(parts) < 5:
                    print(f"Warning: Invalid label format in {label_path}")
                    continue
                    
                class_id = parts[0]
                
                # Skip if class_id is not in our mapping
                if class_id not in class_map:
                    print(f"Warning: Unknown class ID {class_id} in {label_path}")
                    continue
                
                class_name = class_map[class_id]
                x_c, y_c, w, h = map(float, parts[1:5])
                
                # Convert YOLO format to absolute coordinates
                x1 = (x_c - w / 2)
                y1 = (y_c - h / 2)
                x2 = (x_c + w / 2)
                y2 = (y_c + h / 2)
                
                # Make sure coordinates are within image boundaries
                x1 = max(0, min(1, x1)) 
                y1 = max(0, min(1, y1))
                x2 = max(0, min(1, x2))
                y2 = max(0, min(1, y2))
                
                # Convert to pixel coordinates
                xmin = int(x1 * width)
                ymin = int(y1 * height)
                xmax = int(x2 * width)
                ymax = int(y2 * height)
                
                # Skip if bounding box is too small
                if xmax <= xmin or ymax <= ymin:
                    print(f"Warning: Invalid bounding box in {label_path}")
                    continue
                
                # Crop the image
                roi = img[ymin:ymax, xmin:xmax]
                
                # Skip if the ROI is empty
                if roi.size == 0:
                    print(f"Warning: Empty ROI in {image_path} with box {xmin}, {ymin}, {xmax}, {ymax}")
                    continue
                
                # Resize to 32x32
                roi = cv2.resize(roi, (32,32))
                
                # Save with sequential name (6 digits with leading zeros)
                output_name = f"{class_counters[class_name]:06d}.jpg"
                output_path = os.path.join(output_base_path, dataset_type, class_name, output_name)
                cv2.imwrite(output_path, roi)
                
                # Increment counter for this class
                class_counters[class_name] += 1
    
    # Print summary of processed images
    total_images = sum(class_counters.values())
    print(f"Processing complete. {total_images} cropped images saved:")
    for class_name, count in class_counters.items():
        print(f"  - {class_name}: {count} images")

# Define paths
input_base_path = "dataset"
output_base_path = "processed_dataset_32"
    
# Process the dataset
process_dataset(input_base_path, output_base_path)
```

## Analyse Dataset

```python
def get_dataframe(dataset_path):
    
    # Initialize dictionaries to store counts
    train_counts = {}
    val_counts = {}
    
    # Get counts for training set
    train_path = os.path.join(dataset_path, 'train')
    for class_name in os.listdir(train_path):
        class_dir = os.path.join(train_path, class_name)
        if os.path.isdir(class_dir):
            num_images = len([f for f in os.listdir(class_dir) if f.endswith('.jpg')])
            train_counts[class_name] = num_images
    
    # Get counts for validation set
    val_path = os.path.join(dataset_path, 'val')
    for class_name in os.listdir(val_path):
        class_dir = os.path.join(val_path, class_name)
        if os.path.isdir(class_dir):
            num_images = len([f for f in os.listdir(class_dir) if f.endswith('.jpg')])
            val_counts[class_name] = num_images
    
    # Convert to DataFrame for easier plotting
    classes = sorted(list(set(list(train_counts.keys()) + list(val_counts.keys()))))
    
    data = {
        'Class': [],
        'Count': [],
        'Set': []
    }
    
    for cls in classes:
        if cls in train_counts:
            data['Class'].append(cls)
            data['Count'].append(train_counts[cls])
            data['Set'].append('Training')
        
        if cls in val_counts:
            data['Class'].append(cls)
            data['Count'].append(val_counts[cls])
            data['Set'].append('Validation')
    
    df = pd.DataFrame(data)
    
    return df
```

### Visualise the class distribution in dataset

```python
path = "processed_dataset_32"
df = get_dataframe(path)

# Calculate total counts per class (combining train and val)
class_totals = df.groupby('Class')['Count'].sum().reset_index()

# Set up the style
plt.style.use('Solarize_Light2')

plt.figure(figsize=(12, 6))

# Create the grouped bar chart
plt.subplot(1, 2, 1)
chart = sns.barplot(x='Class', y='Count', hue='Set', data=df, palette='viridis')
chart.set_title('Class Distribution in Dataset', fontsize=14)
chart.set_xlabel('Class', fontsize=12)
chart.set_ylabel('Number of Images', fontsize=12)

# Add text labels on the bars
for p in chart.patches:
    height = p.get_height()
    chart.text(p.get_x() + p.get_width()/2.,
            height + 5,
            '{:d}'.format(int(height)),
            ha="center", fontsize=10)

# Create pie chart
plt.subplot(1, 2, 2)
plt.pie(
    class_totals['Count'], 
    labels=class_totals['Class'],
    autopct='%1.1f%%',  # Show percentage with 1 decimal place
    startangle=90, 
    colors=sns.color_palette('viridis', len(class_totals)),
    shadow=False,
    textprops={'fontsize': 12}
)

plt.axis('equal')
plt.title('Overall Class Distribution', fontsize=16)
plt.legend(title="Classes", loc="center left", bbox_to_anchor=(1, 0.5))
plt.tight_layout()

# Display statistics
total_images = class_totals['Count'].sum()
print(f"Total images: {total_images}")

for i, row in class_totals.iterrows():
    cls = row['Class']
    count = row['Count']
    percentage = (count / total_images) * 100
    print(f"{cls}: {count} images ({percentage:.1f}%)")
```

The bar chart and pie chart show the class distribution in the dataset, the result indicates that there is imbalancing in classes. \
In order to address this issue, we consider using `oversampling technique` to increase the number of images for lower classes.

## Simple Oversampling

Here, we duplicate the number of images for `tread_water` class and triplicate `drowning` class

```python
def oversample_images(input_dataset_path, output_dataset_path):
    # Create output directories
    for dataset_type in ['train', 'val']:
        for class_name in ['swimming', 'tread_water', 'drowning']:
            os.makedirs(os.path.join(output_dataset_path, dataset_type, class_name), exist_ok=True)
    
    # Define multiplication factors for each class
    class_multipliers = {
        'swimming': 1,     # Keep as is
        'tread_water': 2,  # Duplicate (2x total)
        'drowning': 3      # Triplicate (3x total)
    }
    
    # Process each dataset type
    for dataset_type in ['train', 'val']:
        print(f"Processing {dataset_type} dataset...")
        
        # Process each class
        for class_name, multiplier in class_multipliers.items():
            class_path = os.path.join(input_dataset_path, dataset_type, class_name)
            output_class_path = os.path.join(output_dataset_path, dataset_type, class_name)
            
            # Skip if directory doesn't exist
            if not os.path.exists(class_path):
                print(f"Warning: Directory not found: {class_path}")
                continue
            
            # Get all images in this class
            image_paths = glob.glob(os.path.join(class_path, '*.jpg'))
            
            if not image_paths:
                print(f"Warning: No images found in {class_path}")
                continue
                
            print(f"  {class_name}: Found {len(image_paths)} images, multiplying by {multiplier}x")
            
            # Copy original images and create duplicates
            counter = 0
            for img_idx, img_path in enumerate(tqdm(image_paths)):
                img_name = os.path.basename(img_path)
                
                # Copy the original image with a new sequential name
                new_img_name = f"{counter:06d}.jpg"
                shutil.copy2(img_path, os.path.join(output_class_path, new_img_name))
                counter += 1
                
                # Create duplicates if needed
                for dup_idx in range(1, multiplier):
                    new_img_name = f"{counter:06d}.jpg"
                    shutil.copy2(img_path, os.path.join(output_class_path, new_img_name))
                    counter += 1
            
            print(f"  {class_name}: Created {counter} images in output directory")
    
    # Print summary and class distribution
    print("\nOversampling complete.")
    print("New class distribution:")
    
    total_counts = {}
    for dataset_type in ['train', 'val']:
        for class_name in class_multipliers.keys():
            class_path = os.path.join(output_dataset_path, dataset_type, class_name)
            if os.path.exists(class_path):
                count = len(glob.glob(os.path.join(class_path, '*.jpg')))
                if class_name not in total_counts:
                    total_counts[class_name] = {}
                total_counts[class_name][dataset_type] = count
    
    # Display table format
    print("\nClass\tTrain\tValidation\tTotal")
    print("-" * 40)
    for class_name in total_counts:
        train_count = total_counts[class_name].get('train', 0)
        val_count = total_counts[class_name].get('val', 0)
        total = train_count + val_count
        print(f"{class_name}\t{train_count}\t{val_count}\t\t{total}")

input_dataset_path = "processed_dataset_32"
output_dataset_path = "balanced_dataset_32"

oversample_images(input_dataset_path, output_dataset_path)
```

## Save Dataset to CSV file

```python
def save_dataset_to_csv(root_dir, output_csv):
    class_mapping = {"swimming": 0, "tread_water": 1, "drowning": 2}
    dataset = []
    
    for split in ["train", "val"]:
        for class_name, class_id in class_mapping.items():
            class_dir = os.path.join(root_dir, split, class_name)
            if os.path.exists(class_dir):
                for filename in os.listdir(class_dir):
                    file_path = os.path.join(class_dir, filename)
                    dataset.append([file_path, class_id])
    
    with open(output_csv, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["file_path", "class_id"])
        writer.writerows(dataset)
    
    print(f"Dataset saved to {output_csv}")

# Usage
save_dataset_to_csv("balanced_dataset_32", "drowning_dataset_32.csv")
```
